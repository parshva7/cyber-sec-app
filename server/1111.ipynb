```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_phishing.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
class_counts = df['status'].value_counts()
plt.figure(figsize=(6,6))
class_counts.plot.pie(autopct='%1.1f%%', colors=['cyan', 'red'])
plt.title('Class Distribution of Dataset (Phishing vs Legitimate)')
plt.ylabel('')
plt.show()
```

```python
df['url_length'] = df['url'].apply(lambda x: len(str(x)))
plt.figure(figsize=(8,6))
plt.hist(df['url_length'], bins=50, color='purple', alpha=0.7)
plt.xlabel('URL Length')
plt.ylabel('Frequency')
plt.title('Distribution of URL Length in Dataset')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Model names and accuracies including XGBoost
model_names = ["Random Forest", "Logistic Regression", "XGBoost"]
accuracies = [0.773, 0.698, 0.783]

# Plotting accuracy bar chart
plt.figure(figsize=(8,5))
bars = plt.bar(model_names, accuracies, color=["green", "blue", "orange"])
plt.ylim(0, 1)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")

# Add accuracy values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.3f}', ha='center', va='bottom')

plt.show()
```
